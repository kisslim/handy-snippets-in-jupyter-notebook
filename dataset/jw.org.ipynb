```python
# crawl all bible like https://www.jw.org/en/library/bible/bi12/books/deuteronomy/33/#v5033001-v5033029

!pip install requests
!pip install tqdm
!pip install bs4
import requests
import tqdm
from bs4 import BeautifulSoup
import re

url = "https://www.jw.org/en/library/bible/bi12/books/deuteronomy/33/#v5033001-v5033029"
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

for a in soup.find_all("a", href=re.compile("/en/library/bible/bi12/")):
    print(a["href"])
```
